```python

spark.sql("""
SELECT company_id, parent_company_id, start_date
FROM (
    SELECT company_df.company_id, company_df.parent_company_id, company_df.start_date
    FROM company_df
    JOIN
    (
        SELECT company_id, count(1) as company_parent_count FROM company_df GROUP BY company_id
    ) single_pc
    on company_df.company_id = single_pc.company_id
    WHERE single_pc.company_parent_count>1
)
WHERE company_id in (
    SELECT DISTINCT company_id FROM log_df
)

EXCEPT

SELECT company_id, parent_company_id, start_date FROM log_df

order by company_id
""").show()
```
