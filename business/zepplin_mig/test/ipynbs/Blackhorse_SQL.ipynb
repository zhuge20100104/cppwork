```python

path_store_daily = "s3://b2c-prod-data-pipeline-unified-store-paid/unified/app-tech.store.app-est-category-rank.v1/fact/granularity=daily/"
df = spark.read.parquet(path_store_daily)
df.createOrReplaceTempView("store_est")
spark.sql("""
select app_id, country_code, device_code, category_id, '2019-06-29' date
from 
(
    select
        app_id, country_code, device_code, count(distinct date) as ranked_date, category_id
    from store_est
    where date >='2019-06-23' and date <='2019-06-29' and category_id in ('400025', '400001', '100021', '100001') and IF(category_id in
    ('100021', '400025'),free_app_download, revenue) <= 100 and device_code in ("android-all", "ios-all") and country_code in ("WW", "US", "CN", "JP")
    group by country_code, app_id, device_code, category_id
) as black_horse_df
where ranked_date>=3
order by app_id, category_id, device_code, country_code
""").show(2000)
```

```python

spark.sql("""
select black_horse_df.app_id
from (
select
    app_id, country_code, device_code, count(distinct date) as ranked_date, category_id
from store_est
where date >='2019-10-01' and date <='2019-10-07' and IF(category_id in ('100021', '400025'),free_app_download, revenue) <= 100
group by country_code, app_id, device_code, category_id
) as black_horse_df 
join(select store_est.app_id from store_est
where date >='2019-10-01' and date <='2019-10-07') as store_est1
on black_horse_df.app_id = store_est1.app_id 
where ranked_date >= 3
""").show()
```
